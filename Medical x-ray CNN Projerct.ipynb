```python
import os

dataset_path = r"C:\Users\hp\Desktop\CNN Project\archive"
print(os.listdir(dataset_path))
```

```python
train_path = r"C:\Users\hp\Desktop\CNN Project\chest_xray\train"
test_path  = r"C:\Users\hp\Desktop\CNN Project\chest_xray\test"
val_path   = r"C:\Users\hp\Desktop\CNN Project\chest_xray\val"
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen  = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_path,
    target_size=(224,224),
    batch_size=32,
    class_mode='binary'
)

val_generator = test_datagen.flow_from_directory(
    val_path,
    target_size=(224,224),
    batch_size=32,
    class_mode='binary'
)

test_generator = test_datagen.flow_from_directory(
    test_path,
    target_size=(224,224),
    batch_size=32,
    class_mode='binary'
)
```

```python
from tensorflow.keras import layers, models

model = models.Sequential([
    
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')  # Binary classification
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10
)
```

```python
test_loss, test_acc = model.evaluate(test_generator)
print("Test Accuracy:", test_acc)
```

```python
model.save("xray_cnn_model.h5")
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

def predict_xray(img_path):
    img = image.load_img(img_path, target_size=(224,224))
    img_array = image.img_to_array(img)/255.0
    img_array = np.expand_dims(img_array, axis=0)

    prediction = model.predict(img_array)

    if prediction[0][0] > 0.5:
        return "PNEUMONIA"
    else:
        return "NORMAL"
```

```python
result = predict_xray(r"C:\Users\hp\Desktop\CNN Project\chest_xray\cheast.png")

print("Prediction:", result)
```

```python
def generate_summary(label):
    if label == "PNEUMONIA":
        return """The X-ray shows signs of lung infection.
        Possible pneumonia detected.
        Medical consultation recommended."""
    else:
        return """The X-ray appears normal.
        No visible signs of infection detected."""
```
